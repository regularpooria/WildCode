```python
#those have two category with the sane probability, both categories are selected
def get_top_categories_by_score(user_categories):
    """
    Returns all category names that share the top probability (rounded to 2 decimal places).
    """
    if not isinstance(user_categories, list) or not user_categories:
        return []

    # Round scores to 2 decimals and find max
    rounded = [(cat, round(prob, 2)) for cat, prob in user_categories]
    max_score = max(score for _, score in rounded)

    # Return all categories with the max score
    return [cat for cat, score in rounded if score == max_score]

# Apply to DataFrame
df["top_categories"] = df["user_init_category"].apply(get_top_categories_by_score)

# Flatten the category list for counting
all_top_cats = [cat for sublist in df["top_categories"] for cat in sublist]
category_counts = Counter(all_top_cats)

# Sort and plot
sorted_counts = dict(sorted(category_counts.items(), key=lambda x: x[1], reverse=True))

plt.figure(figsize=(10, 6))
plt.bar(sorted_counts.keys(), sorted_counts.values())
plt.xlabel("Top Categories (Rounded Prob Ties Allowed)")
plt.ylabel("Frequency")
plt.title("Most Frequent Top Categories by Rounded Probability")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
